```python
import shutil
import os

rt_detr = "/kaggle/working/RT-DETRv3"
output_dir = os.path.join(rt_detr, "output")
os.makedirs(output_dir, exist_ok=True)

# Source checkpoint files
src_dir = "/kaggle/input/preprocess2"
files_to_copy = ["23.pdparams", "23.pdopt", "23.pdema", "23.pdstates"]

# Copy them into output_dir
for f in files_to_copy:
    shutil.copy(os.path.join(src_dir, f), output_dir)

print("Copied:", files_to_copy, "to", output_dir)
```

```python
%%writefile /kaggle/working/RT-DETRv3/configs/datasets/mydata.yml
metric: COCO
num_classes: 1011  

TrainDataset:
  !COCODataSet
    image_dir: images/train
    anno_path: /kaggle/input/preprocess2/train.json
    dataset_dir: /kaggle/input/dataset-with-coco/dataset
    data_fields: ['image', 'gt_bbox', 'gt_class']

EvalDataset:
  !COCODataSet
    image_dir: images/valid
    anno_path: /kaggle/input/preprocess2/val.json
    dataset_dir: /kaggle/input/dataset-with-coco/dataset

TestDataset:
  !ImageFolder
    anno_path: /kaggle/input/preprocess2/test.json
    dataset_dir: /kaggle/input/dataset-with-coco/dataset
    use_default_label: true
```

```python
%%writefile /kaggle/working/RT-DETRv3/configs/rtdetrv3/rtdetrv3_r34vd_6x_coco.yml
_BASE_: [
  '../datasets/mydata.yml',
  '../runtime.yml',
  '_base_/optimizer_6x.yml',
  '_base_/rtdetrv3_r50vd.yml',
  '_base_/rtdetr_reader.yml',
]

weights: output/rtdetrv3_r34vd_6x_coco/model_final
find_unused_parameters: True
log_iter: 200

o2m_branch: True
num_queries_o2m: 450

pretrain_weights: /kaggle/working/ResNet34_vd_pretrained.pdparams

train:
  gradient_accumulation_steps: 4

RTDETRV3:
  backbone: ResNet
  neck: HybridEncoder
  transformer: RTDETRTransformerv3
  detr_head: DINOv3Head
  aux_o2m_head: PPYOLOEHead
  post_process: DETRPostProcess

ResNet:
  depth: 34
  variant: d
  return_idx: [1, 2, 3]
  freeze_at: -1
  freeze_norm: false
  norm_decay: 0.

HybridEncoder:
  hidden_dim: 256
  use_encoder_idx: [2]
  num_encoder_layers: 1
  encoder_layer:
    name: TransformerLayer
    d_model: 256
    nhead: 8
    dim_feedforward: 1024
    dropout: 0.
    activation: 'gelu'
  expansion: 0.5
  depth_mult: 1.0

RTDETRTransformerv3:
  eval_idx: -1
  num_decoder_layers: 4
  num_noises: 3
  num_noise_queries: [300, 300, 300]
  num_noise_denoising: 100
  learnt_init_query: False
```

```python
%%writefile /kaggle/working/RT-DETRv3/configs/rtdetrv3/_base_/rtdetr_reader.yml
worker_num: 4
TrainReader:
  sample_transforms:
    - Decode: {}
    - RandomDistort: {prob: 0.8}
    - RandomExpand: {fill_value: [123.675, 116.28, 103.53]}
    - RandomCrop: {prob: 0.8}
    - RandomFlip: {}
  batch_transforms:
    - BatchRandomResize: {target_size: [480, 512, 544, 576, 608, 640, 640, 640, 672, 704, 736, 768, 800], random_size: True, random_interp: True, keep_ratio: False}
    - NormalizeImage: {mean: [0., 0., 0.], std: [1., 1., 1.], norm_type: none}
    - NormalizeBox: {retain_origin_box: true}
    - BboxXYXY2XYWH: {}
    - Permute: {}
    - PadGT: {only_origin_box: true}
  batch_size: 4
  shuffle: true
  drop_last: true
  collate_batch: false
  use_shared_memory: true


EvalReader:
  sample_transforms:
    - Decode: {}
    - Resize: {target_size: [640, 640], keep_ratio: False, interp: 2}
    - NormalizeImage: {mean: [0., 0., 0.], std: [1., 1., 1.], norm_type: none}
    - Permute: {}
  batch_size: 2
  shuffle: false
  drop_last: false


TestReader:
  inputs_def:
    image_shape: [3, 640, 640]
  sample_transforms:
    - Decode: {}
    - Resize: {target_size: [640, 640], keep_ratio: False, interp: 2}
    - NormalizeImage: {mean: [0., 0., 0.], std: [1., 1., 1.], norm_type: none}
    - Permute: {}
  batch_size: 1
  shuffle: false
  drop_last: false
```
